```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_numbers = np.random.uniform(0, 1, 1000)
plt.hist(random_numbers, bins=100, edgecolor='black')

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("1000 Uniform Numbers")

plt.savefig("uniform_histogram.pdf")
plt.show()
```
