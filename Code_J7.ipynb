```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0, 1, 100)
def my_exp(x):
    return np.exp(x)
y = my_exp(x)
plt.plot(x, y)

plt.ylabel("Awesomeness")
plt.xlabel("Time [milliseconds]")
plt.title("Growth of Awesomeness Over Time")

plt.savefig("awesomeness_plot.pdf")
plt.show()
```
