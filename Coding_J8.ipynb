```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)

def s_sin(x):
    return np.sin(x)

def c_cos(x):
    return np.cos(x)
```

```python
y_sin = s_sin(x)
y_cos = c_cos(x)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.plot(x, y_sin)
ax1.set_title("sin(x)")
ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax2.plot(x, y_cos)
ax2.set_title("cos(x)")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
plt.tight_layout()

plt.savefig("sin_cos_plot.pdf")
plt.show()
```
